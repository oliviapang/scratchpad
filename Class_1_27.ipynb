```python
x = df['Make_Year']
q75 = np.quantile(x, 0.75)
q25 = np.quantile(x, 0.25)
iqr = q75-q25
print(iqr)
uw = q75+1.5*iqr
lw = q25-1.5*iqr

upper_outlier = (x > uw).astype(int)
print(upper_outlier)
```
