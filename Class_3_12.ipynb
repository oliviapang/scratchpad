```python
y=df['TARGET_LABEL_BAD=1']
x = df.drop('TARGET_LABEL_BAD=1', axis=1)
```

```python
from sklearn.linear_model import LogisticRegression

def maxmin(z):
  u = (z-min(z))/(max(z)-min(z))
  return u

U = x.apply(maxmin)

model = LogisticRegression(penalty=None)
model = model.fit(U,y)
p_hat = model.predict_proba(U)
```
